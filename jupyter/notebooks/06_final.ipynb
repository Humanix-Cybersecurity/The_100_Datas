### Le code maître
Les 5 fragments t'ouvrent le mainframe de CASSANDRA. Il ne reste qu'une commande
à envoyer pour suspendre le protocole. Elle dit littéralement ce que tu dois faire.

> Arrête. L'. Abattage. → en anglais, la langue du système : **STOP THE CULLING**

🎯 **Fragment final :** `ARK{STOP_THE_CULLING}` · valide-le dans la Console.

### 🏆 Débrief · la vraie leçon
Vous avez gagné non pas en piratant l'IA, mais en **auditant ses données**.

- Fallait-il croire CASSANDRA *parce que* c'était une IA ?
- Qui est responsable d'un Abattage décidé par un algorithme : la machine,
  son concepteur, le Conseil qui l'a déployée sans audit ?
- Que faudrait-il pour qu'une IA de ce type soit *auditable* et *équitable* ?
